Pre-processing MNIST data:<br>

Get the MNIST dataset ready for training by scaling its pixel intensities and reshaping each image so it includes one channel. Scaling the values makes training more stable and helps the model converge sooner, while the reshape step is needed because the diffusion model’s input expects 3D image-shaped data.
